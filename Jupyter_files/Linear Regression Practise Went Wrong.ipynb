```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('car_dataset.csv')
```

```python
df
```

```python
df = df.drop(['Engine Cylinders','Number of Doors','Engine Fuel Type'],axis =1)
```

```python
df.head()
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
label = le.fit_transform(df['Transmission Type'])
label
```

```python
df.drop(['Transmission Type'],axis =1,inplace = True)
```

```python
df['Transmission Type'] = label
df
```

```python
df = df.dropna()
x = df.drop(['MSRP','Make','Model','Driven_Wheels','Market Category','Vehicle Size','Vehicle Style'],axis =1)
y = df['MSRP']
x
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(x,y)
```

```python
y_prediction = log_reg.predict(x)
y_prediction
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y,y_prediction))
```
